Importar librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import emoji
```

Carga del dataset de train

```python
df_train = pd.read_csv('../data/irosva.mx.training.csv')
print(df_train.shape)
print(df_train.head())
```

```python
print(df_train.columns.tolist())
print(df_train.dtypes)
```

```python
print(df_train['IS_IRONIC'].value_counts())
print()
print(df_train['IS_IRONIC'].value_counts(normalize=True) * 100)
```

```python
df_train['longitud'] = df_train['MESSAGE'].apply(len)
print(df_train['longitud'].describe())
```

```python
print("=== TWEETS IRÓNICOS ===")
for tweet in df_train[df_train['IS_IRONIC']==1]['MESSAGE'].head(5).values:
    print("-", tweet)

print("\n=== TWEETS NO IRÓNICOS ===")
for tweet in df_train[df_train['IS_IRONIC']==0]['MESSAGE'].head(5).values:
    print("-", tweet)
```

```python
def tiene_emoji(texto):
    return any(char in emoji.EMOJI_DATA for char in texto)

df_train['tiene_emoji'] = df_train['MESSAGE'].apply(tiene_emoji)
print(df_train.groupby('IS_IRONIC')['tiene_emoji'].mean() * 100)
```

```python
df_mx = pd.read_csv('../data/irosva.mx.training.csv')
df_es = pd.read_csv('../data/irosva.es.training.csv')
df_cu = pd.read_csv('../data/irosva.cu.training.csv')
```

```python
for nombre, df in [('México', df_mx), ('España', df_es), ('Cuba', df_cu)]:
    total = len(df)
    ironico = df['IS_IRONIC'].sum()
    no_ironico = total - ironico
    print(f"{nombre}: Total={total} | Irónico={ironico} ({ironico/total*100:.1f}%) | No irónico={no_ironico} ({no_ironico/total*100:.1f}%)")
```

```python
for nombre, df in [('México', df_mx), ('España', df_es), ('Cuba', df_cu)]:
    df['tiene_emoji'] = df['MESSAGE'].apply(tiene_emoji)
    por_clase = df.groupby('IS_IRONIC')['tiene_emoji'].mean() * 100
    print(f"\n{nombre}:")
    print(f"  No irónico: {por_clase[0]:.1f}% con emoji")
    print(f"  Irónico:    {por_clase[1]:.1f}% con emoji")
```
